```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
```

```python
distortions = []
inertias = []
mapping1 = {}
mapping2 = {}
KK = range(1, 15)
 
for k in KK:
    # Building and fitting the model
    kmeanModel = KMeans(n_clusters=k).fit(df)
    kmeanModel.fit(df)
 
    distortions.append(sum(np.min(cdist(df, kmeanModel.cluster_centers_,
                                        'euclidean'), axis=1)) / df.shape[0])
    inertias.append(kmeanModel.inertia_)
 
    mapping1[k] = sum(np.min(cdist(df, kmeanModel.cluster_centers_,
                                   'euclidean'), axis=1)) / df.shape[0]
    mapping2[k] = kmeanModel.inertia_
```

```python

for key, val in mapping1.items():
    print(f'{key} : {val}')
```

```python
plt.plot(KK, distortions, 'bx-')
plt.xlabel('Liczba klastrów')
plt.ylabel('Zniekształcenie')
plt.title('Metoda łokcia dla zniekształceń')
plt.show()
```

```python

for key, val in mapping2.items():
    print(f'{key} : {val}')
```

```python

plt.plot(KK, inertias, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertion')
plt.title('Elbow method for inertion')
plt.show()
```
